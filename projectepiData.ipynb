```python
# Tracer  pour montrer la distribution des buts sur les saisons et les ann√©es civiles
fig, axes = plt.subplots(2, 1, figsize = (25,25))

# The Distribution of Goals among the Seasons (Bar Chart)
seasons = events.Season.value_counts().sort_index()
axes[0].bar(seasons.index, seasons.values, color = ['#6ee1fa', '#65d4ee', '#54b9d7', '#5cc6e3', '#449fbf', '#2d7a9b', '#1f6282',
                                                    '#003653', '#07415f', '#3c93b3', '#104c6b', '#3c93b3', '#175776', '#3586a7',
                                                    '#266e8f', '#4caccb', '#449fbf'])
axes[0].set_title('Goals Distribution over Seasons', size = 36)
axes[0].set_xlabel('Season', size = 24)
axes[0].set_ylabel('Goals', size = 24)
axes[0].patch.set_facecolor('#F2F2F2')
axes[0].tick_params(labelrotation = 75, labelsize = 20)
axes[0].spines['bottom'].set_color('black')
axes[0].spines['left'].set_color('black')
axes[0].spines['top'].set_color('black')
axes[0].spines['right'].set_color('black')

# The Distribution of Goals among the calendar Years (Bar Chart)
years = events.Year.value_counts().sort_index()
axes[1].bar(years.index, years.values, color = ['#6ee1fa', '#65d4ee', '#54b9d7', '#5cc6e3', '#3c93b3', '#07415f', '#104c6b',
                                                '#003653', '#3586a7', '#1f6282', '#266e8f', '#175776', '#1f6282', '#266e8f',
                                                '#2d7a9b', '#4caccb', '#449fbf'])
axes[1].set_title('Goals Distribution over Calender Years', size = 36)
axes[1].set_xlabel('Year', size = 24)
axes[1].set_ylabel('Goals', size = 24)
axes[1].patch.set_facecolor('#F2F2F2')
axes[1].tick_params(labelrotation = 75, labelsize = 20)
axes[1].locator_params('x', nbins = 34)
axes[1].spines['bottom'].set_color('black')
axes[1].spines['left'].set_color('black')
axes[1].spines['top'].set_color('black')
axes[1].spines['right'].set_color('black')
```

Fonction qui sera utilis√©e


```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,8))
fig.suptitle('Distribution of the Goals in the competitions', size = 36)

sns.countplot(events['Competition'], ax = axes[0], palette = "rocket");
axes[0].set_xlabel('Competition', size=24)
axes[0].set_ylabel('No. of Goals', size=24)
axes[0].tick_params(labelrotation = 90, labelsize = 20)
axes[0].spines['bottom'].set_color('black')
axes[0].spines['left'].set_color('black')
axes[0].spines['top'].set_color('black')
axes[0].spines['right'].set_color('black')

competition = events.Competition.value_counts().sort_values(ascending = False)
axes[1].pie(competition.values, explode = (0.1, 0.2, 0.3, 0.5, 0.7, 0.9), labels = competition.index, shadow = True,
            colors = ['#08737f', '#00898a', '#089f8f', '#39b48e', '#64c987', '#92dc7e'],
            textprops = {'family':'sans-serif', 'fontsize':'large', 'fontstyle':'italic', 'fontweight':'heavy'},
            autopct = lambda pct: func(pct, competition.values));
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,8))

UCL_Goals = pd.Series({'Final': 2, 'Semi-Finals': 6, 'Quarter-Finals': 12, 'Last 16': 29, 'Group Stage': 71})
axes[0].pie(UCL_Goals.values, labels = UCL_Goals.index, explode = (0.5, 0.3, 0.1, 0.05, 0.05), shadow = True,
            autopct = lambda pct: func(pct, UCL_Goals.values),
            colors = ['#c4ec74', '#64c987', '#089f8f', '#08737f', '#2a4858'],
            textprops = {'family':'sans-serif', 'fontsize':'large', 'fontstyle':'italic', 'fontweight':'heavy'});
axes[0].set_title('UEFA Champions League', fontsize = 20);


Copa_Goals = pd.Series({'Final': 9, 'Semi-Finals': 11, 'Quarter-Finals': 11, 'Last 16': 22, 'Others': 3})
axes[1].pie(Copa_Goals.values, labels = Copa_Goals.index, explode = (0.05, 0.05, 0.05, 0.05, 0.05), shadow = True,
            autopct = lambda pct: func(pct, Copa_Goals.values),
            colors = ['#39b48e', '#00898a', '#00898a', '#215d6e', '#fafa6e'],
            textprops = {'family':'sans-serif', 'fontsize':'large', 'fontstyle':'italic', 'fontweight':'heavy'});
axes[1].set_title('Copa del Rey', fontsize = 20);
```

```python
Place = events.Venue.value_counts()
fig, ax = plt.subplots(figsize = (12, 8))
ax.pie(Place.values, labels = ['Home', 'Away'], shadow = True, autopct = lambda pct: func(pct, Place.values),
       colors = ['#08737f', '#92dc7e'],
       textprops = {'family':'sans-serif', 'fontsize':'x-large', 'fontstyle':'italic', 'fontweight':'heavy'});
plt.title('Distribution of the Goals between Home and Away', size = 24)
plt.legend(['Home', 'Away'])
plt.show()
```
